```python
# Uncomment the following lines if running in a new environment
# %pip install pandas numpy matplotlib seaborn scikit-learn fairlearn shap openpyxl

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from fairlearn.metrics import MetricFrame
# import shap  # Uncomment after installing shap
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load and examine MEPS data
df = pd.read_csv('h233.csv')
pd.set_option('display.max_columns', None)
print(df.head(10))
print(f"Shape: {df.shape}")
print("Columns:", df.columns[:25].tolist())

# Column selection and renaming
selected_columns = [
    'AGE21X', 'SEX', 'RACEV1X', 'HISPANX', 'FAMINC21',
    'EDUCYR', 'INSURC21', 'RTHLTH53', 'MNHLTH53',
    'OBTOTV21', 'ERTOT21', 'IPNGTD21', 'TOTEXP21'
]
df_small = df[selected_columns].copy().rename(columns={
    'AGE21X': 'age',
    'SEX': 'sex',
    'RACEV1X': 'race',
    'HISPANX': 'hispanic',
    'FAMINC21': 'income',
    'EDUCYR': 'education_years',
    'INSURC21': 'insured',
    'RTHLTH53': 'self_rated_health',
    'MNHLTH53': 'self_mental_health',
    'OBTOTV21': 'office_visits',
    'ERTOT21': 'er_visits',
    'IPNGTD21': 'inpatient_nights',
    'TOTEXP21': 'total_expenditure'
})
df_small = df_small.dropna()

for col in ['sex', 'race', 'hispanic', 'insured']:
    df_small[col] = df_small[col].astype('category')

df_model = pd.get_dummies(df_small, columns=['sex', 'race', 'hispanic', 'insured'], drop_first=True)

print("Race category codes:", dict(enumerate(df_small['race'].cat.categories)))
```

```python
all_features = [col for col in df_model.columns if col != 'total_expenditure']
proxy_cols = ['office_visits', 'er_visits', 'inpatient_nights'] + [col for col in df_model.columns if col.startswith('insured_')]

X_with_proxy = df_model[all_features]
features_no_proxy = [f for f in all_features if f not in proxy_cols]
X_no_proxy = df_model[features_no_proxy]
y = df_model['total_expenditure']
```

```python
X_train_wp, X_test_wp, y_train, y_test = train_test_split(X_with_proxy, y, test_size=0.2, random_state=42)
X_train_np = X_no_proxy.loc[X_train_wp.index]
X_test_np = X_no_proxy.loc[X_test_wp.index]
```

```python
param_dist = {
    'n_estimators': [100, 200, 500],
    'max_depth': [5, 10, 20, None],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

search = RandomizedSearchCV(
    RandomForestRegressor(random_state=42),
    param_distributions=param_dist,
    n_iter=10, cv=3, scoring='neg_mean_squared_error', n_jobs=-1
)
search.fit(X_train_np, y_train)
print('Best hyperparameters (no-proxies model):', search.best_params_)
rf_np = search.best_estimator_
```

```python
# With proxies
rf_wp = RandomForestRegressor(random_state=42)
rf_wp.fit(X_train_wp, y_train)
y_pred_wp = rf_wp.predict(X_test_wp)

# Without proxies (optimized)
y_pred_np = rf_np.predict(X_test_np)

print("\nWith proxies:")
print('  MSE:', mean_squared_error(y_test, y_pred_wp))
print('  R^2:', r2_score(y_test, y_pred_wp))
print("\nWithout proxies (optimized):")
print('  MSE:', mean_squared_error(y_test, y_pred_np))
print('  R^2:', r2_score(y_test, y_pred_np))
```

```python
test_results = X_test_wp.copy()
test_results['y_true'] = y_test
test_results['pred_with_proxy'] = y_pred_wp
test_results['pred_no_proxy'] = y_pred_np

race_columns = [col for col in test_results.columns if col.startswith("race_")]
group_results = []
for race_var in race_columns:
    subgroup = test_results[test_results[race_var] == 1]
    if not subgroup.empty:
        mse_proxy = mean_squared_error(subgroup['y_true'], subgroup['pred_with_proxy'])
        mse_noproxy = mean_squared_error(subgroup['y_true'], subgroup['pred_no_proxy'])
        group_results.append({'group': race_var, 'mse_proxy': mse_proxy, 'mse_noproxy': mse_noproxy})

group_df = pd.DataFrame(group_results)
if not group_df.empty:
    group_df.set_index('group')[['mse_proxy', 'mse_noproxy']].plot(kind='bar')
    plt.title("Group MSE with vs. without proxy features")
    plt.ylabel("MSE")
    
    
    plt.show()
```

```python
black_col_name = 'race_2'  # Adjust based on your category mapping!
if black_col_name in X_train_wp.columns:
    sample_weights = np.where(X_train_wp[black_col_name] == 1, 2.0, 1.0)
    rf_weighted = RandomForestRegressor(random_state=42)
    rf_weighted.fit(X_train_wp, y_train, sample_weight=sample_weights)
    y_pred_weighted = rf_weighted.predict(X_test_wp)
    print('Weighted model (race_2 upweighted): MSE:', mean_squared_error(y_test, y_pred_weighted))
else:
    print(f"Column {black_col_name} not found in features!")
```

```python
weighted_test_results = X_test_wp.copy()
weighted_test_results['y_true'] = y_test
weighted_test_results['pred_weighted'] = y_pred_weighted

mse_weighted_by_group = {}
mse_unweighted_by_group = {}
for race_var in race_columns:
    subw = weighted_test_results[weighted_test_results[race_var] == 1]
    subu = test_results[test_results[race_var] == 1]
    if not subw.empty and not subu.empty:
        mse_weighted_by_group[race_var] = mean_squared_error(subw['y_true'], subw['pred_weighted'])
        mse_unweighted_by_group[race_var] = mean_squared_error(subu['y_true'], subu['pred_with_proxy'])

mse_comparison_df = pd.DataFrame({
    'Unweighted MSE': pd.Series(mse_unweighted_by_group),
    'Weighted MSE': pd.Series(mse_weighted_by_group)
}).sort_index()

print('\nGroup-wise comparison of MSE (Unweighted vs Weighted):')
print(mse_comparison_df)

mse_comparison_df.plot(kind='bar', figsize=(10,6))
plt.title('Group-wise MSE Comparison: Weighted vs Unweighted Models')
plt.ylabel('Mean Squared Error')
plt.xlabel('Race Group')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Residuals by group
test_results['residuals_unweighted'] = test_results['y_true'] - test_results['pred_with_proxy']
weighted_test_results['residuals_weighted'] = weighted_test_results['y_true'] - weighted_test_results['pred_weighted']

for race_var in race_columns:
    mean_resid_unw = test_results.loc[test_results[race_var]==1, 'residuals_unweighted'].mean()
    mean_resid_w = weighted_test_results.loc[weighted_test_results[race_var]==1, 'residuals_weighted'].mean()
    print(f"{race_var} mean residual (unweighted): {mean_resid_unw:.2f}, mean residual (weighted): {mean_resid_w:.2f}")

# Feature importances plot
importances = rf_np.feature_importances_
feat_imp = pd.DataFrame({'feature': X_no_proxy.columns, 'importance': importances})
feat_imp.sort_values('importance', ascending=False).head(10).plot.bar(x='feature', y='importance', title='Top 10 Features')
plt.show()

# SHAP summary (optional, uncomment if shap is installed)
# explainer = shap.TreeExplainer(rf_np)
# shap_values = explainer.shap_values(X_test_np)
# shap.summary_plot(shap_values, X_test_np)
```

```python
proxy_X = df_model[proxy_cols]
race_all = df_small['race'].cat.codes
X_proxy_train = proxy_X.loc[X_train_wp.index]
X_proxy_test = proxy_X.loc[X_test_wp.index]
yrace_train = race_all.loc[X_train_wp.index]
yrace_test = race_all.loc[X_test_wp.index]

proxy_clf = LogisticRegression(max_iter=500, multi_class='multinomial')
proxy_clf.fit(X_proxy_train, yrace_train)
yrace_pred = proxy_clf.predict(X_proxy_test)
random_chance = 1 / len(np.unique(yrace_test))
proxy_acc = accuracy_score(yrace_test, yrace_pred)

print("\n=== Proxy Detection: Predicting Race from Proxies (test set) ===")
print(f"Proxy classifier accuracy (random chance is {random_chance:.1%}): {proxy_acc:.2%}")
print("Classification Report:")
print(classification_report(yrace_test, yrace_pred))
```
